```python
import pandas as pd
import glob 
```

```python
files = glob.glob("markers/*.csv")
```

```python
m_markers = "|".join('''Plxdc2
Elmo1
Srgap2
Tanc2
Inpp5d
Hexb
Nav2
Tgfbr1
Zfhx3
Cd11b
Cx3cr1
Tmem119'''.split("\n"))
m_markers
```

```python
dfs = []
for i, file in enumerate(files):
    df = pd.read_csv(files[0], index_col=0)
    df = df.reset_index()
    df = df.sort_values(by="p_val_adj").reset_index(drop=True)
    df = df[df["index"].str.contains(m_markers, case=False)]
    df["cluster"] = i+1
    df.insert(0, "gene", df.pop("index"))
    df = df.reset_index()
    dfs.append(df)
    
```

```python
df = pd.concat(dfs)
```

```python
df = df.sort_values(by=["index","cluster"])
df
```
